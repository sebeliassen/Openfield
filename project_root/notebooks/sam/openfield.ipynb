```python
import pandas as pd
import numpy as np
import os
```

```python
def process_session(session, freq, led_state, sampling_rate=15):
    phot_df = session.dfs.data["phot"]
    filtered_phot_df = phot_df[phot_df["LedState"] == led_state]
    comp_timestamp_idx = filtered_phot_df.columns.get_loc("ComputerTimestamp")
    data_columns = filtered_phot_df.columns[comp_timestamp_idx + 1:]

    for data_column in data_columns:
        brain_region, side, _ = data_column
        data = filtered_phot_df[data_column]
        timestamp = filtered_phot_df["SystemTimestamp"]

        guppy_df = pd.DataFrame({
            'timestamps': timestamp,
            'data': data
        })
        guppy_df.reset_index(inplace=True, drop=True)
        guppy_df["sampling_rate"] = np.nan
        guppy_df.at[0, "sampling_rate"] = sampling_rate

        trial_dirs = session.trial_dir.split('/')
        prefix = '/'.join(trial_dirs[:-1]) + '_guppy/' + trial_dirs[-1]

        if not os.path.exists(prefix):
            os.makedirs(prefix)

        file_name = f'{prefix}/guppy_{session.mouse_id}_{freq}_{brain_region}_{side}.csv'
        guppy_df.to_csv(file_name, index=False)

for session in sessions:
    for freq, led_state in zip([415, 470], [1, 2]):
        process_session(session, freq, led_state)
#phot_df_470 = phot_df[phot_df["LedState"] == 2]
```
